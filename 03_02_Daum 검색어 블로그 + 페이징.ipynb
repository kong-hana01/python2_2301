## Daum 검색어 블로그 + 페이징
1. url 분석

https://search.daum.net/search?w=blog&DA=PGD&enc=utf8&q=%ED%8C%8C%EC%9D%B4%EC%8D%AC&p=55

```python
from bs4 import BeautifulSoup
import requests
import urllib.parse as urlparser

def crowling_keyword_in_daum_blog(keyword, page):
    url = f"https://search.daum.net/search?w=blog&nil_search=btn&DA=NTB&enc=utf8&q={urlparser.quote(keyword)}&p={page}"
    response = requests.get(url)
    data = BeautifulSoup(response.text, "html.parser")
    blog = []
    for element in data.select(".f_l a.f_nb")[::2]:
        blog.append(element.text.strip())
    title = []
    for element in data.select(".f_link_b"):
        title.append(element.text.strip())
    url = []
    for element in data.select(".f_link_b"):
        url.append(element.attrs['href'].strip())
    result = []
    for i in range(len(title)):
        temp = {"post_title": title[i], "blog_name": blog[i], "url": url[i]}
        result.append(temp)
    return result
```

```python
start_page, end_page = map(int, input("시작 페이지와 끝 페이지를 입력해주세요.").split())
for i in range(start_page, end_page+1):
    print(f'{i}번째 다음 블로그 크롤링 중')
    print(crowling_keyword_in_daum_blog("파이썬", i))
    print()
```
