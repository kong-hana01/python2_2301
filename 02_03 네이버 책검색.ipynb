## 네이버 책 검색
### urlparser
query string을 살펴보면 검색한 내용이 브라우저 url 창에서는 한글로 되어있지만 복사 붙여넣기를 하면 %xx의 조합으로 이뤄진 것을 확인할 수 있습니다.<br>
한글은 그대로 쿼리에서 사용할 수 없기에 url encoding을 해야하기 때문입니다.<br>
ex) https://search.shopping.naver.com/search/all?query=%ED%8C%8C%EC%9D%B4%EC%8D%AC%EC%B1%85&cat_id=&frm=NVSHATC

```python
from bs4 import BeautifulSoup
import requests
import urllib.parse as urlparser # url encoding

keyword = '파이썬책'
print(keyword)
print(urlparser.quote(keyword))
```

## 완성 코드

```python
from bs4 import BeautifulSoup
import requests
import urllib.parse as urlparser # url encoding

def crowling_shoping_in_naver(keyword):
    url = f'https://search.shopping.naver.com/search/all?query={urlparser.quote(keyword)}'
    response = requests.get(url)
    data = BeautifulSoup(response.text, "html.parser")

    title = []
    for element in data.select(".basicList_link__JLQJf"):
        title.append(element.text.strip())
    price = []
    for element in data.select(".price_num__S2p_v"):
        price.append(element.text.strip())
    url = []
    for element in data.select(".basicList_link__JLQJf"):
        url.append(element.attrs['href'])
    result = []
    for i in range(len(title)):
        temp = {"책 제목": title[i], "가격": price[i], "상세 페이지": url[i]}
        result.append(temp)
    print(result)
```

```python
crowling_shoping_in_naver("파이썬")
```
