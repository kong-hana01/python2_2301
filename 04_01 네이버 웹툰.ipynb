url 분석

https://comic.naver.com/webtoon/detail?titleId=783056&no=121&weekday=wed

```python
import requests
from bs4 import BeautifulSoup
url = "https://comic.naver.com/webtoon/detail?titleId=783056&no=121&weekday=wed"
response = requests.get(url)
print(response)
```

```python
data = BeautifulSoup(response.text, "html.parser")
print(data)
```

```python
img_elements = data.select(".wt_viewer img")
img_srcs = []
for element in img_elements:
    img_srcs.append(element.attrs['src'])
```

### 이미지 한 장 다운로드

```python
img_src = img_srcs[0]
print(img_src)
```

```python
response = requests.get(img_src, stream=True)
print(response) # 권한 에러
```

```python
# header -> user-agent에 크롬 정보를 기입한 뒤, requests
user_agent = "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/109.0.0.0 Safari/537.36"
headers = {"user-agent": user_agent}
```

```python
response = requests.get(img_src, stream = True, headers = headers)
print(response)
```

```python
if response.status_code == 200:
    with open("여고생드래곤.jpg", "wb") as f:
        f.write(response.content)
```

### 완성 코드

```python
import requests
from bs4 import BeautifulSoup
import os
dir_path = "./img/"
if not os.path.isdir(dir_path): # 폴더가 없다면,
    os.mkdir(dir_path)          # 폴더를 생성하세요!
    
url = "https://comic.naver.com/webtoon/detail?titleId=783056&no=121&weekday=wed"
response = requests.get(url)
data = BeautifulSoup(response.text, "html.parser")
img_elements = data.select(".wt_viewer img")
img_srcs = []
for element in img_elements:
    img_srcs.append(element.attrs['src'])
user_agent = "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/109.0.0.0 Safari/537.36"
headers = {"user-agent": user_agent}

for i in range(len(img_srcs)):
    img_src = img_srcs[i]
    response = requests.get(img_src, stream=True, headers=headers) # 각 이미지 파일에 대해 requests
    if response.status_code == 200:
        with open(f'./img/{i+1}.jpg', "wb") as f:
            f.write(response.content)
```
